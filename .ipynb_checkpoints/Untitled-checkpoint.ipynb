```python
with open('红海行动.txt',encoding='utf-8') as f:
    rows = f.readlines()
```

```python
rows[0]
```

```python
i =1
boxes = []
box = []
for row in rows:
    if 'div' in row:
        continue
    if row.strip()=='--':
        row = 'None'
    box.append(row.strip())
    if i%13 == 0 :
        boxes.append(box)
        box=[]
    i+=1
```

```python
boxes[:5]
```

```python
import csv
with open('红海行动.csv', 'w', newline='',encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(boxes)
```

```python
# import sklearn 
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('红海行动.csv')
data2 = data
data2.iloc[:,[1,3,5,6,8,11]] = data2.iloc[:,[1,3,5,6,8,11]].applymap(lambda x : x if  '%' not in x else x[:-1])
sits = data['座位数']
sits2=sits.apply(lambda x : float(x[:-1])*10000 if '万' in x else x)
data2['座位数'] = sits2
data2['场次']=data['场次'].map(lambda x : float(x[:-1])*10000 if '万' in x else x)
data2 = data2.convert_objects(convert_numeric=True)

data2.head()
```

```python
Y_train = data2['综合票房']
X_train = data2.drop(['分账票房','综合票房','分账票房占比','场均人次','网售占比','综合票房占比','综合票价'],axis = 1)
X_train['holiday']=0
X_train['weekday']=0

for i in range(X_train.shape[0]):
    if i%7 == 1 or i%7 ==2:
        X_train.iloc[i,6] = 1
    else:
        X_train.iloc[i,7] = 1
X_train.head()
```

```python
X_train['time'] = 0
for i in range(X_train.shape[0]):
    X_train.iloc[i,8] = np.sqrt(1.0/(i+1))
X_train.head()
```

```python
X_train.to_csv('hhxd_train_x.csv',index = False)
Y_train.to_csv('hhxd_train_y.csv',index = False)
```
