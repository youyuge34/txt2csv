```python
# import sklearn 
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('zzl2.csv')
```

```python
data.applymap(lambda x: x).head(25)
```

```python
data2 = data
data2.iloc[:,[1,3,5,6,8,11]] = data2.iloc[:,[1,3,5,6,8,11]].applymap(lambda x : x if  '%' not in x else x[:-1])
data2.head()
```

```python
sits = data['座位数']
sits2=sits.apply(lambda x : float(x[:-1])*10000 if '万' in x else x)
data2['座位数'] = sits2
data2['场次']=data['场次'].map(lambda x : float(x[:-1])*10000 if '万' in x else x)
data2.head()
```

```python
data2 = data2.convert_objects(convert_numeric=True)
data2.info()
```

```python
Y_train = data2['综合票房']
X_train = data2.drop(['分账票房','综合票房'],axis = 1)
```

```python
Y_train.head()
```

```python
X_train = X_train.drop(['分账票房占比','场均人次','网售占比','综合票房占比','综合票价'],axis = 1)
X_train.shape
```

```python
X_train['holiday']=0
```

```python
X_train['weekday']=0
```

```python
for i in range(X_train.shape[0]):
    if i%7 == 2 or i%7 ==3:
        X_train.iloc[i,6] = 1
    else:
        X_train.iloc[i,7] = 1
X_train.head()
```

```python
X_train['time'] = 0
for i in range(X_train.shape[0]):
    X_train.iloc[i,8] = np.sqrt(1.0/(i+1))
X_train.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score


ss_x = StandardScaler()
ss_y = StandardScaler()
```

```python
X_data,X_test,Y_data,Y_test = train_test_split(X_train.iloc[:50],Y_train.iloc[:50],test_size=0.3,shuffle=True)
```

```python
X_data = ss_x.fit_transform(X_data)
X_test = ss_x.transform(X_test)
Y_data = ss_y.fit_transform(Y_data.values.reshape(-1,1))
Y_test = ss_y.transform(Y_test.values.reshape(-1,1))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_data,Y_data)
lr_y_predict = lr.predict(X_data)
print(lr.score(X_data,Y_data))
print(lr.score(X_test,Y_test))
print(ss_y.inverse_transform(lr.predict(X_test)))
print(ss_y.inverse_transform(Y_test))
```

```python
# plt.subplot(111)
plt.figure(figsize=(8,5))
total = X_data.shape[0]+X_test.shape[0]
plt.plot(range(total),Y_train.iloc[:total],'--',color='seagreen')
plt.plot(range(total),ss_y.inverse_transform(lr.predict(ss_x.transform(X_train.iloc[:total]))))
plt.show()
```

```python
lr.coef_
```

```python
from sklearn.svm import SVR

svr = SVR(kernel='rbf')
svr.fit(X_data,Y_data)
print(svr.score(X_data,Y_data))
print(svr.score(X_test,Y_test))
print(ss_y.inverse_transform(svr.predict(X_test)))
print(ss_y.inverse_transform(Y_test))
```

```python
# plt.subplot(111)
plt.figure(figsize=(8,5))
total = X_data.shape[0]+X_test.shape[0]
plt.plot(range(total),Y_train.iloc[:total],'--',color='seagreen')
plt.plot(range(total),ss_y.inverse_transform(svr.predict(ss_x.transform(X_train.iloc[:total]))))
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor()
rfr.fit(X_data,Y_data)
print(rfr.score(X_data,Y_data))
print(rfr.score(X_test,Y_test))
print(ss_y.inverse_transform(rfr.predict(X_test)))
print(ss_y.inverse_transform(Y_test))
```
